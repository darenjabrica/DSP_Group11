```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000)

# Define the input signal x(t) as a sine wave (representing sound pressure)
# Frequency of 1 Hz for simplicity
x_t = np.sin(2 * np.pi * t)

# 2. System Equation: Amplification by factor of 2
y_t = 2 * x_t

# 3. Shifting the Input
tau = 2  # Time shift (2 seconds)
x_shifted = np.sin(2 * np.pi * (t - tau))  # Shifted input signal x(t - tau)
y_shifted = 2 * x_shifted  # Output for the shifted input

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot original input signal
plt.subplot(2, 1, 1)
plt.plot(t, x_t, label='Original Input x(t)')
plt.plot(t, y_t, label='Amplified Output y(t)=2*x(t)', linestyle='--', color = 'red')
plt.title('Original Signal and Amplified Output')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()

# Plot shifted signal
plt.subplot(2, 1, 2)
plt.plot(t, x_shifted, label='Shifted Input x(t - tau)')
plt.plot(t, y_shifted, label='Amplified Shifted Output y(t - tau)=2*x(t - tau)', linestyle='--', color = 'red')
plt.title('Shifted Input and Amplified Shifted Output')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
t = np.linspace(0, 10, 1000)

# Define the electricity usage x(t) as a sine wave (cyclic usage)
x_t = np.sin(t)

# 2. System Equation: Multiplication of time t with usage x(t)
y_t = t * x_t

# 3. Shifting the Input
tau = 2  # Time shift (2 hours)
x_shifted = np.sin(t - tau)  # Shifted input signal x(t - tau)
y_shifted = (t - tau) * x_shifted  # Output for the shifted input

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot original input signal
plt.subplot(2, 1, 1)
plt.plot(t, x_t, label='Electricity Usage x(t)')
plt.plot(t, y_t, label='Cost of Electricity y(t)=t*x(t)', linestyle='--')
plt.title('Original Usage and Cost of Electricity')
plt.xlabel('Time (hours)')
plt.ylabel('Amplitude')
plt.legend()

# Plot shifted signal
plt.subplot(2, 1, 2)
plt.plot(t, x_shifted, label='Shifted Usage x(t - tau)')
plt.plot(t, y_shifted, label='Shifted Cost y(t - tau)=(t - tau)*x(t - tau)', linestyle='--')
plt.title('Shifted Usage and Cost of Electricity')
plt.xlabel('Time (hours)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
t = np.linspace(0, 10, 1000)

# Define the car's position x(t) as a sine wave
x_t = np.sin(t)

# 2. System Equation: Differentiation of the car's position x(t) to get velocity
dt = t[1] - t[0]  # Calculate time step
y_t = np.gradient(x_t, dt)  # Numerical derivative of x(t)

# 3. Shifting the Input
tau = 2  # Time shift (2 seconds)
x_shifted = np.sin(t - tau)  # Shifted position signal x(t - )
y_shifted = np.gradient(x_shifted, dt)  # Velocity for the shifted input

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot original position and velocity
plt.subplot(2, 1, 1)
plt.plot(t, x_t, label='Car Position x(t)')
plt.plot(t, y_t, label='Velocity y(t)=dx/dt', linestyle='--')
plt.title('Original Car Position and Velocity')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()

# Plot shifted position and velocity
plt.subplot(2, 1, 2)
plt.plot(t, x_shifted, label='Shifted Position x(t - tau)')
plt.plot(t, y_shifted, label='Shifted Velocity y(t - tau)=d/dt[x(t - tau)]', linestyle='--')
plt.title('Shifted Car Position and Velocity')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
