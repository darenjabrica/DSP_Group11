```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p1():
    n = np.arange(0, 100)
    cs = np.sin(2 * np.pi * n / 25)
    noise = np.random.normal(0, 0.5, cs.shape)
    ns = cs + noise

    fs = np.zeros_like(ns)
    
    for i in range(2, len(ns)):
        fs[i] = (ns[i] + ns[i-1] + ns[i-2]) / 3

    plt.figure(figsize=(10, 4))
    plt.plot(n, ns, label='Noisy Signal')
    plt.plot(n, fs, label='Filtered Signal', linestyle='--', color='green')
    plt.title('Digital Filtering in Signal Processing')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

p1()
```

```python
def p2():
    n = np.arange(0, 100)
    bp = 100
    ss = bp + np.sin(2 * np.pi * n / 10) + np.random.normal(0, 1, 100)
    
    sm = np.zeros_like(ss)
    
    for i in range(1, len(ss) - 1):
        sm[i] = (ss[i-1] + ss[i] + ss[i+1]) / 3
    
    plt.figure(figsize=(10, 4))
    plt.plot(n, ss, label='Original Stock Prices')
    plt.plot(n, sm, label='Smoothed Stock Prices', linestyle='--')
    plt.title('Stock Price Prediction')
    plt.xlabel('Sample Index')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

p2()
```

```python
def p3():
    n = np.arange(0, 100)
    cs = np.sin(2 * np.pi * n / 25)
    
    d = 5
    a = 0.6
    es = np.zeros_like(cs)
    es[d:] = a * cs[:-d]
    
    swe = cs + es
    
    canc_s = np.copy(swe)
    for i in range(d, len(swe)):
        canc_s[i] = swe[i] - a * cs[i - d]

    plt.figure(figsize=(10, 4))
    plt.plot(n, cs, label='Original Signal')
    plt.plot(n, swe, label='Signal with Echo', linestyle='--')
    plt.plot(n, canc_s, label='Echo Canceled Signal', linestyle=':', color='red')
    plt.title('Echo Cancellation in Telecommunications')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

p3()
```
