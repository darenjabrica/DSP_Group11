```python
import numpy as np
import matplotlib.pyplot as plt 

def f(t):
    y = np.exp(-t/2)
    return y
```

```python
## CODE 1

x =  np.linspace(0,10,1000)
plt.plot (x, f(x))
plt.show()
```

```python
## CODE 2
T = np.arange(0, 8)
Y = np.array([0, 1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3])

plt.plot(T, Y)
plt.show()
```

```python
##CODE 3
f = np.sin
N_low = np.linspace(0, 2*np.pi, 5) 
plt.stem(N_low, f(N_low))
plt.plot(N_low, f(N_low), color='blue')

plt.show()

```

```python
##CODE 4
N_high = np.linspace(0, 2*np.pi, int(1e6))  

plt.stem(N_high, f(N_high))
plt.plot(N_high, f(N_high), color='blue')

plt.show()
```

```python
##CODE 5
N = np.linspace(0, 2*np.pi, int(4*np.pi))  
plt.stem(N, f(N))
plt.plot(N, f(N), color='blue')

plt.show()
```
