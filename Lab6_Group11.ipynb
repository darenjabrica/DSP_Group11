```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_sam = 100

n = np.arange(n_sam)
x = np.sin(0.2 * np.pi * n)

lti = np.zeros(n_sam)  

for i in range(1, n_sam):
    lti[i] = 0.7 * lti[i-1] + x[i]

plt.figure(figsize=(10, 5))
plt.plot(n, x, label='Input Signal x[n] = sin(0.2pi*n)', color='b')
plt.plot(n, lti, label='LTI System Output y[n]', color='r')
plt.title('LTI System Response')
plt.xlabel('n (Sample Number)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_nl = x ** 2


plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.scatter(n, x, color='b', label='Input Signal x[n] = sin(0.2pi*n)')
plt.title('Non-Linear System Response - Input Signal (Scatter Plot)')
plt.xlabel('n (Sample Number)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.bar(n, y_nl, color='r', label='Non-Linear System Output y[n] = x[n]^2')
plt.title('Non-Linear System Response - Output Signal (Bar Plot)')
plt.xlabel('n (Sample Number)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
x_cl = np.sin(0.3 * np.pi * n)

noise = np.random.normal(0, 0.2, n_sam)  # Mean = 0, Std Dev = 0.2
x_n = x_cl + noise

y_n = np.zeros(n_sam)

for i in range(1, n_sam):
    y_n[i] = 0.5 * y_n[i-1] + 0.5 * x_n[i]

plt.figure(figsize=(12, 6))

plt.plot(n, x_cl, label='Clean Input Signal x_cl[n] = sin(0.3pi*n)', color='b')

plt.plot(n, x_n, label='Noisy Input Signal', color='green')

plt.plot(n, y_n, label='System Output y[n]', color='r')

plt.title('LTI System Response to Noisy Input Signal')
plt.xlabel('n (Sample Number)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
